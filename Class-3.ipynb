```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
df['age']=df['age'].fillna(df.age.median())
df
```

```python
df['deck']=df['deck'].cat.add_categories('unknown')
df['deck']=df['deck'].fillna('unknown')
```

```python
df.shape
```

```python
df['age_group']=pd.cut(df['age'],bins=range(0,95,10))
surv_by_age=df.groupby('age_group')['survived'].mean()
surv_by_age.plot(kind='line',marker='o',figsize=(5,5))
plt.title("Survival Rate By Age")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
plt.scatter(df['age'],df['fare'])
plt.title("Age And Fare Distribution")
plt.ylabel("Fare")
plt.xlabel("Age")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,4))
sc=ax.scatter(df['age'],df['fare'],c=df['survived'],cmap='coolwarm')
ax.set_title('Scatter Plot ------ Age Vs Fare')
ax.set_xlabel("Age")
ax.set_ylabel('Fare (Dollars)')
cbar=fig.colorbar(sc,ax=ax,ticks=[0,1])
cbar.ax.set_yticklabels(['Died','Survived'])
plt.show()
```

```python
df['class'].value_counts().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('People')
plt.title('Class Wise People')
plt.show()
```

```python
df['embark_town'].value_counts().plot(kind='barh')
plt.title('Embark_town Count')
plt.ylabel('Town')
plt.xlabel('Count')
plt.show()
```

```python
grp=df.groupby(['class','sex'])['survived'].mean().unstack()
x=np.arange(len(grp.index))
w=0.35
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(x-w/2,grp['male'],w,label="Male")
plt.bar(x+w/2,grp['female'],w,label="Female")
plt.xticks(x,grp.index)
plt.title("Survived rate")
plt.legend()
stack=df.groupby(['class','survived']).size().unstack()
plt.subplot(1,2,2)
plt.bar(stack.index,stack[0],label='Died')
plt.bar(stack.index,stack[1],bottom=stack[0],label='Survived')
plt.title("Outcome Counts")
plt.legend()
```

```python
plt.figure(figsize=(5,4))
plt.hist(df['age'],bins=25)
plt.title("Age Distibution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
plt.savefig('first_plot',)
```

```python
fig, ax=plt.subplots(2,2,figsize=(8,8))

ax[0,0].hist(df['age'],bins=20,color="skyblue")
ax[0,0].set_title("Age Distribution")

ax[0,1].hist(df['fare'],bins=20,color='orange')
ax[0,1].set_title("Fare Distribution")

class_count=df['class'].value_counts()
ax[1,0].bar(class_count.index,class_count.values,color='green')
ax[1,0].set_title("Passenger By Class")

survival=df.groupby('sex')['survived'].mean()
ax[1,1].bar(survival.index,survival.values,color='pink')
ax[1,1].set_title("Survival rate By Gender")

plt.suptitle("Titanic dashboard")
plt.tight_layout()
plt.show()
```

```python
cols=['survived','pclass','age','sibsp','fare']
corr=df[cols].corr()
corr
plt.figure(figsize=(5,4))
sns.heatmap(corr,annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title("Feature Correlattion Map")
plt.show()
```

```python
df['age_group']=pd.cut(df['age'],bins=range(0,85,5))
tmp = (
    df.dropna(subset=['age_group'])
      .assign(age_mid=lambda x: x['age_group'].map(lambda i: i.mid))
)
sns.lineplot(data=tmp,x='age_mid',y='survived',hue='sex',markers='o',errorbar=('ci',95))
```

 ### Histplot

```python
sns.histplot(data=df,x='age',hue='survived',multiple='stack')
plt.title("Age Vs Survival Rate")
```

### Kdeplot

```python
sns.kdeplot(data=df,x='fare',hue='class',fill=True,alpha=0.4, common_norm=False,clip=(0,200))
plt.title("Fare vs Class")
```

```python
fig,ax=plt.subplots(1,2,figsize=(9,3))
sns.boxplot(data=df,x='class',y='fare',hue='survived',palette='Set2',ax=ax[0])
ax[0].set_title("Boc plot fare vs survival ")
sns.violinplot(data=df,x='class',y='age',hue='sex',palette='muted',split=True,ax=ax[1])
ax[1].set_title("Violin plot -age vs class and survival ")
plt.show()
```

### Heat Map


```python
numeric_cols=['survived','pclass','age','sibsp','parch','fare']
corr=df[numeric_cols].corr()
plt.figure(figsize=(5,4))
sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    square=True
)
plt.title('heatmap()= Feature correction')
plt.show()
```

### Cluster Map

```python
corr=df[numeric_cols].corr()
g=sns.clustermap(corr,annot=True,fmt='.2f',cmap='coolwarm',center=0,figsize=(5,5),linewidths=0.5)
g.figure.suptitle("Clustermap()-Clustered Feature Correlation",y=1.02)
plt.show()
```
