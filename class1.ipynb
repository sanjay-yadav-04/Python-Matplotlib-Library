## Starting Matplotlib

```python
# s = [5,10, 15,20, 50]
plt.figure(figsize=(5,3))
plt.hist(df['total_bill'], range = (10, 40))
plt.title("Distribution of total bill")
plt.xlabel('Total Bill')
plt.ylabel("Frequency")
plt.show()
```

```python
avg_tip = df.groupby('day')['tip'].mean()

plt.figure()
avg_tip.plot(kind='bar', colormap = 'cool')
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure()


for size, group in df.groupby('size'):
    count = len(group)
    label_text = f"Size {size} (n={count})"
    
    
    plt.scatter(group['total_bill'], group['tip'], label=label_text)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title="Party Size & Counts")
```

```python
plt.figure()

scatter = plt.scatter(df['total_bill'], df['tip'], c = df['size'], cmap='coolwarm')

cbar = plt.colorbar(scatter)

cbar.set_label('Party Size')
```

```python
plt.Figure()
df.boxplot(column='total_bill',by='time')
plt.title("Total Bill By Time")
plt.ylabel('Total Bill')
plt.xlabel('Total Time')
plt.show()
```

```python
a=df.groupby('day')['tip'].mean()
a
plt.Figure()
plt.plot(a,marker='o')
plt.title("Average Tip By Day")
plt.xlabel('day')
plt.ylabel('tip')
plt.show()
```

```python
cust_count=df.groupby('day')['size'].sum()
plt.figure()
plt.bar(cust_count.index,cust_count.values)
plt.title("Customer Count by day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
smoker_count=df['smoker'].value_counts()

plt.figure(figsize=(3,3))
plt.pie(smoker_count.values, labels= 
    smoker_count.index, autopct='%1.1f%%',shadow=True)
plt.title("Smoker vs non-smoker")
plt.show()
```
